```python
# pip install PyMuPDF

import fitz

def pdf_content_extractor(file_path):
    pdf_document = fitz.open(file_path)
    
    for page_number in range(len(pdf_document)):
        page = pdf_document.load_page(page_number)
        page_text = page.get_text("text")
        
        lines = page_text.split('\n')
        filtered_lines = [line for line in lines if line.strip() != "This output was created with the assistance of Generative AI technology."]
    
        filtered_text = '\n'.join(filtered_lines)
        
        print(f"Page {page_number + 1}:\n{filtered_text}\n")


file_path = r"C:\Users\hichoudhary\Downloads\KDD_kddex 1218 chp_24273.pdf"
pdf_content_extractor(file_path)
```

```python
from docx import Document

def initialize_extracted_data():
    return {
        "KDD Title": "",
        "KDD ID": "",
        "Date Opened": "",
        "Approved Date": "",
        "Owner": "",
        "Status": "",
        "Key Stakeholders / Approvers": "",
        "KDD Description": "",
        "Impacted Process": "",
        "Impacted Details": "",
        "Impacted Function Assumptions": "",
        "Impacted Function Details": "",
        "Solutions": [],
        "Merits": [],
        "Demerits": [],
        "Recommendation": ""
    }


def word_extraction(file):
    doc = Document(file)  
    extracted_data = initialize_extracted_data() 
    current_section = None

    for table in doc.tables:
        for row in table.rows:
            cells = row.cells
            key = cells[0].text.strip()
            value = cells[1].text.strip()

            if key in extracted_data:
                extracted_data[key] = value

 
    for para in doc.paragraphs:
        text = para.text.strip()

        if text.lower().startswith("solutions"):
            current_section = "Solutions"
        elif text.lower().startswith("merits"):
            current_section = "Merits"
        elif text.lower().startswith("demerits"):
            current_section = "Demerits"
        elif text.lower().startswith("recommendation"):
            current_section = "Recommendation"
        elif text == "":
            current_section = None
        else:
            
            if current_section == "Solutions":
                extracted_data["Solutions"].append(text)
            elif current_section == "Merits":
                extracted_data["Merits"].append(text)
            elif current_section == "Demerits":
                extracted_data["Demerits"].append(text)
            elif current_section == "Recommendation":
                extracted_data["Recommendation"] += " " + text
    
    return extracted_data  


file_path = 'KDD_cx seg kdd 03_24342.docx'
extracted_info = word_extraction(file_path)


for key, value in extracted_info.items():
    if isinstance(value, list):
        print(f"{key}:")
        for item in value:
            print(f"  - {item}")
    else:
        print(f"{key}: {value}")
```
